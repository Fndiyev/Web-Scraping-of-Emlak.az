```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.yeniemlak.az/arenda-obyektler-icareye'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find all pagination links
pagination_links = soup.select("a[href*='page']")  # Adjust selector

all_pages = set()
for link in pagination_links:
    page_url = link["href"]
    if page_url.startswith("/"):
        page_url = "https://example.com" + page_url
    all_pages.add(page_url)

print(all_pages)  # This will give you all available pages
```
